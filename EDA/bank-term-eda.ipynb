```python
import seaborn as sns
import matplotlib.pyplot as plt

jobs = train['job'].value_counts()
plt.figure(figsize=(15,8))
sns.barplot(x=jobs.index, y=jobs.values, palette='Set1')
plt.xlabel('Jobs', fontsize=20)
plt.ylabel('Job Count', fontsize=20)
plt.title('Job Distribution', fontsize=20)
plt.show()
```

```python
marital_counts = train['marital'].value_counts()

plt.figure(figsize=(10,6))
plt.pie(
    marital_counts.values,              
    labels=marital_counts.index,        
    autopct='%1.1f%%',                   
    startangle=160,
    explode=[0.05, 0.05, 0.05],
    shadow=True                     
    )

plt.title('Marital Status Distribution',  fontsize=20)
plt.axis('equal') 
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(data=train, x='age', bins=20, kde=True)
plt.title('Age Distribution', fontsize=20)
plt.xlabel('Age', fontsize=20)
plt.ylabel('Count', fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(
    data=train,
    x='marital',        
    hue='education', palette='Set2'    
)
plt.title('Marital Status by Education Level', fontsize=20)
plt.xlabel('Marital Status', fontsize=15)
plt.ylabel('Count', fontsize=15)
plt.xticks(rotation=45)
plt.legend(title='Education', fontsize=15)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,6))
sns.countplot(
    data=train,
    x='housing',     
    hue='loan',
    palette='Set3'      
)
plt.title('Housing vs Loan Status', fontsize=15)
plt.xlabel('Housing', fontsize=15)
plt.ylabel('Count', fontsize=15)
plt.legend(title='Loan Aproval', fontsize=10)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.violinplot(
    data=train,
    x='education',   
    y='y'           
)
plt.title('Outcomes Distribution by Education Level ', fontsize=15)
plt.xlabel('Education', fontsize=15)
plt.ylabel('Outcomes', fontsize=15)
plt.xticks(rotation=45)
plt.show()
```

```python
monthly_y = train.groupby(['month','y']).size().reset_index(name='count')

plt.figure(figsize=(10,6))
sns.lineplot(data=monthly_y, x='month', y='count', hue='y', style='y', markers=True, dashes=True, palette='Set1')
plt.title('Monthly Records Split by Aproval' , fontsize=15)
plt.xlabel('Month', fontsize=15)
plt.ylabel('Count', fontsize=15)
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='contact', data=train, palette='Set2')
plt.title('Contact Type Distribution', fontsize=15)
plt.xlabel('Contacts Types', fontsize=15)
plt.ylabel('Count of Contacts Types', fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='y', y='balance', data=train, palette='Set2')
plt.title('Balance by Y', fontsize=16, fontweight='bold')
plt.xlabel('Y', fontsize=12)
plt.ylabel('Balance', fontsize=12)
plt.show()
```
